Mapping the three digit zip code to the geographic center of all five digit codes that start with the three digit zip

```python
import numpy as np
import pandas as pd
```

```python
zip5_df=pd.read_csv("C:/Users/krish/Desktop/Preety/us-zip.csv", sep =";")
```

```python
print(zip5_df)
```

```python
zip5_df["zip3"] = [z//100 for z in zip5_df["Zip"]]
```

```python
zip5_df["zip_geo_center"] = [(str(x) + "," + str(y)) for x in zip5_df.groupby("zip3")['Latitude'].transform('mean') for y in zip5_df.groupby("zip3")['Longitude'].transform('mean')]
```

```python
zip5_df["zip_geo_center"] = [str(x) +","+ str(y) for x, y in zip(zip5_df.groupby("zip3")['Latitude'].transform('mean'), zip5_df.groupby("zip3")['Longitude'].transform('mean'))]
```
